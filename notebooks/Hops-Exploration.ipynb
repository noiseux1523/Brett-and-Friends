```python
with open('../docs/Yakima_Chief_Hops_Varieties.txt', 'r') as f:
    hops_pages = []
    page = []
    for line in f:
        if '\x0c' in line:
            hops_pages.append(page)
            page = []
            page.append(line.strip())
        else:
            page.append(line.strip())
```

```python
for hop_page in hops_pages[1:]:
    hop_page
    hop_dict = {'Hop': '', 'Description': '', 'Country': '', 'Aroma': '', 'Beer Style': '', 
                'Alpha Acid': '', 'Alpha Acid - Low': '', 'Alpha Acid - High': '', 'Co-Humulone - Low': '', 'Co-Humulone - High': '', 'Beta Acid - Low': '', 'Beta Acid - High': '', 'Total Oil - Low': '', 'Total Oil - High': '',  
                'B-Pinene': '', 'Myrcene': '', 'Linalool': '', 'Caryophyllene': '', 'Farnesene': '', 'Humulene': '', 'Geraniol': '', 'Selinene': '', 'Other': ''}
    # Hop Name
    for line in hop_page:
        if '\x0c' in line:
            hop_dict['Hop'] = line.strip('\x0c')
            
    # Country
    country = []
    country_tag = False
    for line in hop_page:
        if line == 'COUNTRY':
            country_tag = True
        elif country_tag:
            country.append(line)
        if line == '':
            country_tag = False
    hop_dict['Country'] = ' '.join(country)
    
    # Country
    aroma = []
    aroma_tag = False
    for line in hop_page:
        if line == 'AROMA PROFILE':
            aroma_tag = True
        elif aroma_tag:
            aroma.append(line)
        if line == '':
            aroma_tag = False
    hop_dict['Aroma'] = aroma[0].split(' • ')
    
    # Beer style
    style = []
    style_tag = False
    for line in hop_page:
        if line == 'BEER STYLES':
            style_tag = True
        elif style_tag:
            style.append(line)
        if line == '':
            style_tag = False
    hop_dict['Beer Style'] = style[0].split(' • ')
    
    # Humulone
    alpha_acid = []
    alpha_acid_tag = False
    for line in hop_page:
        if 'ALPHA ACIDS' in line:
            hop_dict['Alpha Acid'] = line
    if not hop_dict['Alpha Acid'] == 'ALPHA ACIDS':
        hop_dict['Co-Humulone - Low'] = hop_dict['Alpha Acid'].split('(')[1][:2]
        if len(hop_dict['Alpha Acid'].split(' - ')) > 1:
            hop_dict['Co-Humulone - High'] = hop_dict['Alpha Acid'].split(' - ')[1][:2]
    hop_dict
```
