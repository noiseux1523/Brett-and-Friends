### 1.1 Information Extraction

```python
nlp = spacy.load("en_core_web_sm")
def preprocess(nlp, sent):
    doc = nlp(sent)
    tags = []
    for chunk in doc.noun_chunks:
        print(chunk.text, chunk.root.text, chunk.root.dep_,
                chunk.root.head.text)
    for token in doc:
        print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
                token.shape_, token.is_alpha, token.is_stop, '\n')
        if token.pos_ in ['ADJ', 'VERB', 'NOUN']:
            tags.append(token.lemma_)
    return tags
hops['pos'] = hops.Characteristics.apply(lambda sent: preprocess(nlp, sent) if isinstance(sent, str) else -1)
```

```python
stopwords = ['resonance', 'note', 'same', 'hop', 'aroma', 'use', 'character', 'essence', 'flavor', 'exhibit', 'profile', 'style', 'tasting', 'characteristic']
def filterTokens(sent):
    return [word for word in sent if word not in stopwords]
hops['pos2'] = hops.pos.apply(lambda sent: filterTokens(sent) if isinstance(sent, list) else -1)
hops['pos2'].tolist()
```
